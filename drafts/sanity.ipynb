```python
def prepare_dataset(raw_df):
    raw_df[CT_ACCESSION_COL] = raw_df['StudyInstanceUID'] + '_' + raw_df['SeriesInstanceUID']
    final_df = raw_df.groupby(CT_ACCESSION_COL,as_index=False).sum()[[CT_ACCESSION_COL,'negative_exam_for_pe']]
    final_df["PE"] = (final_df['negative_exam_for_pe']==0).astype(int)
    return final_df.drop(columns=['negative_exam_for_pe']), raw_df
```

```python
final_df, train_raw_df = prepare_dataset(train_raw_df)
final_df[CT_ACCESSION_COL].nunique() , len(final_df)
```
